```python
movies = pd.read_csv("/kaggle/input/letterboxd-movie-ratings-data/movie_data.csv",lineterminator='\n')
```

```python
movies
```

```python
movies.drop(['imdb_id','imdb_link','image_url','_id','spoken_languages',
                 'tmdb_link','overview','tmdb_id','release_date',
                 'production_countries','runtime','vote_count','vote_average','movie_id'],axis=1,inplace=True)
```

```python
movies
```

```python
movies.genres.replace('[]', np.nan, inplace=True)
movies.genres.dropna(inplace=True)
movies.genres.value_counts()
```

```python
import pandas as pd
m = movies.original_language.str.contains('en', na = False)
m
```

```python
finalMovies = movies[m]
finalMovies
```

```python
finalMovies.genres.value_counts()
```

```python
topPopularGenre = finalMovies.genres.value_counts()
tpg = topPopularGenre[:25]
```

```python
import plotly.express as px
fig = px.bar(tpg)
```

```python
fig.show()
```

```python
Docu = finalMovies[finalMovies['genres'].str.contains("(\[\"Documentary\"\])",na=False)]
Docu
```

```python
fg = Docu[:30]
figDocu = px.bar(fg, x = "movie_title", y = "popularity", title = "Documentary Popularity and Year Released", color = "year_released")
figDocu.show()
```

```python
Drama = finalMovies[finalMovies['genres'].str.contains("(\[\"Drama\"\])",na=False)]
Drama
```

```python
fg1 = Drama[:30]
figDrama = px.bar(fg1, x = "movie_title", y = "popularity", title = "Drama Popularity and Year Released", color = "year_released")
figDrama.show()
```

```python
figOverall = px.funnel(finalMovies[:50], x = "genres", y = "year_released", title = "Genres map", color = "popularity")
figOverall.show()
```
