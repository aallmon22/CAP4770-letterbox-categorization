```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Convert genres and movie_titles to numerical representations
genre_mapping = {genre: idx for idx, genre in enumerate(finalMovies['genres'].unique())}
finalMovies['genres'] = finalMovies['genres'].map(genre_mapping)

# Normalize the popularity scores
finalMovies['popularity'] = (finalMovies['popularity'] - finalMovies['popularity'].mean()) / finalMovies['popularity'].std()

# Split the dataset into training and testing sets
train_data, test_data = train_test_split(finalMovies, test_size=0.2)

# Create the neural network model
model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(1,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)
])

# Compile the model with the correct optimizer argument
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(train_data['genres'], train_data['popularity'], epochs=100, batch_size=32, validation_split=0.2)

# Evaluate the model
loss = model.evaluate(test_data['genres'], test_data['popularity'])
print(f'Test Loss: {loss}')

five_years_data = pd.DataFrame({'genres': finalMovies['genres'].unique()})
#ten_years_data = pd.DataFrame({'genres': finalMovies['genres'].unique()})
#twenty_years_data = pd.DataFrame({'genres': finalMovies['genres'].unique()})

# Predict popularity for each time point
five_years_predictions = model.predict(five_years_data['genres'])
#ten_years_predictions = model.predict(ten_years_data['genres'])
#twenty_years_predictions = model.predict(twenty_years_data['genres'])


# EVERYTHING BELOW NEEDS WORK
#five_years_predictions_original_scale = (five_years_predictions * original_popularity_std) + original_popularity_mean

#import matplotlib.pyplot as plt

#genres = ['Genre1', 'Genre2', 'Genre3', ...]
#predicted_popularity = five_years_predictions_original_scale.squeeze()  

#plt.bar(genres, predicted_popularity)
#plt.xlabel('Genres')
#plt.ylabel('Predicted Popularity')
#plt.title('Predicted Popularity of Genres in 5 Years')
#plt.xticks(rotation=45)
#plt.show()
```
